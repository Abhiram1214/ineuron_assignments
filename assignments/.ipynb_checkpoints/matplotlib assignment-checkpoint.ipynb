```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import optimize
```

```python
titanic = pd.read_csv('titanic_original.csv')
titanic.head()
```

## 1. Create a pie chart presenting the male/female proportion

```python
colors = ['magenta', 'pink']
male_female = titanic['sex'].value_counts()
labels = ['Male', 'Female']
```

```python
male_female
```

```python
plt.pie(male_female, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('pie chart presenting the male/female proportion')
plt.show()
```

```python

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

months = np.arange(12)
plt.plot(months, temp_max, 'blue')
plt.plot(months, temp_min, 'green')
plt.title("Min-Max tremparatures")
plt.xlabel('Mnths')
plt.ylabel('Temparature')
plt.show()
```

```python
def yearly_temparature(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temparature, months, temp_max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temparature, months, temp_min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, temp_max, 'go')
plt.plot(days, yearly_temparature(days, *res_max), 'm-')
plt.plot(months, temp_min, 'co')
plt.plot(days, yearly_temparature(days, *res_min), 'y-')
plt.xlabel('Month')
plt.ylabel('Temperature')

plt.show()
```
